```python
# Importing the necessary modules
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
#plt.style.use('seaborn-dark')
plt.rcParams['figure.figsize']=8,8
```

```python
def generate_and_plot(Kx, mu, color = 'lime'):

  distr = multivariate_normal(
    cov = Kx, mean = mu,
    seed = 1000)

  data = distr.rvs(size = 5000)

  plt.plot(data[:,0],data[:,1], 'o', c=color,
             markeredgewidth = 0.5,
             markeredgecolor = 'black')
  plt.grid()

  plt.title(r'Random samples from a 2D-Gaussain distribution')

  plt.xlim(-5, 5)
  plt.ylim(-5, 5)

  plt.xlabel(r'$x_1$')
  plt.ylabel(r'$x_2$')
  plt.axis('equal')
  plt.show()
```

```python
def generate_data(Kx, mu, num = 5000):

  distr = multivariate_normal(
    cov = Kx, mean = mu,
    seed = 1000)

  data = distr.rvs(size = num)

  return data
```

```python
def plot_data(data, color='lime', xlabel = r'$x_1$', ylabel = r'$x_2$' ):

  plt.grid()

  plt.plot(data[:,0],data[:,1], 'o', c=color,
             markeredgewidth = 0.5,
             markeredgecolor = 'black')

  plt.title(r'Random samples from a 2D-Gaussain distribution')
  plt.xlabel(xlabel)
  plt.ylabel(r'$x_2$')
  plt.axis('equal')
```

## Random Vector:  $K_{xx} =  \begin{bmatrix}
  2 & -1.8\\
  -1.8 & 2
  \end{bmatrix}$  and $\mu_x =   \begin{bmatrix}
  1\\
  2\\
  \end{bmatrix}$ .

```python
Kx_1 = np.array([[2.0, -1.8],[-1.8, 2.0]])
mu_1 = np.array([1,2])
random_seed = 10
```

```python
generate_and_plot(Kx_1, mu_1)
```

## Random Vector:  $K_{xx} =  \begin{bmatrix}
  5 & 0.5\\
  0.5 & 0.8
  \end{bmatrix}$  and $\mu_x =   \begin{bmatrix}
  0\\
  -2\\
  \end{bmatrix}$ .

```python
Kx_2 = np.array([[5, 0.5],[0.5, 0.8]])
mu_2 = np.array([0,-2])
random_seed = 10
```

```python
generate_and_plot(Kx_2, mu_2, color ='blue')
```

## Random Vector:  $K_{xx} =  \begin{bmatrix}
  2 & 1.8\\
  1.8 & 2
  \end{bmatrix}$  and $\mu_x =   \begin{bmatrix}
  -2\\
  -3\\
  \end{bmatrix}$ .

```python
Kx_3 = np.array([[2.0, 1.8],[1.8, 2.0]])
mu_3 = np.array([6,6])
random_seed = 10
```

```python
generate_and_plot(Kx_3, mu_3, color ='red')
```

### Gaussian Mixture Model:     $ \ \ \ \  \ p(x|\mu, K) = w_1\ p(x|\mu_1,K_1) + w_2 \ p(x|\mu_2, K_2) + w_3 \ p(x|\mu_3, K_3)$

```python
x1 = generate_data(Kx_1, mu_1, num = 1500)
x2 = generate_data(Kx_2, mu_2, num = 2000)
x3 = generate_data(Kx_3, mu_3, num = 2500)
```

$ \ \ \ \  \ p(x|\mu, K) = \frac{1}{4}\ p(x|\mu_1,K_1) + \frac{1}{3}  \ p(x|\mu_2, K_2) + \frac{5}{12} \ p(x|\mu_3, K_3)$

```python
plt.grid()
plt.plot(x1[:,0],x1[:,1], 'o', c='lime', markeredgewidth = 0.5, markeredgecolor = 'black')
#plt.hold()
plt.plot(x2[:,0],x2[:,1], 'o', c='blue', markeredgewidth = 0.5, markeredgecolor = 'black')
plt.plot(x3[:,0],x3[:,1], 'o', c='red', markeredgewidth = 0.5, markeredgecolor = 'black')
```
